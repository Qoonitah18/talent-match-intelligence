```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Tentukan nama file dan bobot
file_name = "summary_insights terbaru.csv"
df_insights = pd.read_csv(file_name)
df_insights.columns = ['variable', 'category', 'avg_high', 'avg_non_high', 'difference']

# 1. Filter data hanya untuk variabel yang menjadi penentu formula
final_variables = ['gtq', 'STO', 'Papi_V']
df_filtered = df_insights[df_insights['variable'].isin(final_variables)].copy()

# 2. Tambahkan bobot sebagai kolom baru untuk ditampilkan
# Bobot: GTQ (0.448), STO (0.398), Papi_V (0.154)
weight_map = {'gtq': 0.448, 'STO': 0.398, 'Papi_V': 0.154}
df_filtered['weight'] = df_filtered['variable'].map(weight_map)
df_filtered['label'] = df_filtered.apply(lambda row: f"{row['variable']} ({row['weight']:.3f})", axis=1)

# Sortir berdasarkan difference untuk visualisasi (opsional)
df_filtered = df_filtered.sort_values(by='difference', ascending=False)

# --- Membuat Visualisasi Bar Chart (Comparison Matrix) ---
plt.figure(figsize=(10, 6)) # Ukuran figura ditingkatkan
sns.barplot(
    x='difference',
    y='label',
    data=df_filtered,
    palette=['#348AA7', '#5DD39E', '#E87461'] # Warna yang jelas
)

plt.title('Justifikasi Formula: Kesenjangan Kinerja (Gap) Berdasarkan TGV Kunci', fontsize=14)
plt.xlabel('Selisih Skor Median (Gap) - Dasar untuk Bobot TGV', fontsize=12)
plt.ylabel('Variabel Talenta (TV) & Bobot Akhir', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=11) # Ukuran font Y-axis ditingkatkan
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()

# Simpan dan tampilkan plot
plot_filename = 'justifikasi_formula_final.png'
plt.savefig(plot_filename)
print(f"Visualisasi Bar Chart Justifikasi (Comparison Matrix) telah disimpan sebagai {plot_filename}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_name = "summary_insights terbaru.csv"
df_insights = pd.read_csv(file_name)
df_insights.columns = ['variable', 'category', 'avg_high', 'avg_non_high', 'difference']

# Filter data untuk kategori 'competency'
df_comp = df_insights[df_insights['category'] == 'competency'].sort_values(by='difference', ascending=False)

# Membuat Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(
    x='difference',
    y='variable',
    data=df_comp,
    palette='Reds_d'
)
plt.title('1. Kesenjangan Skor: Kategori Kompetensi (Pilihan Formula: STO)', fontsize=14)
plt.xlabel('Selisih Skor (GAP)', fontsize=12)
plt.ylabel('Pilar Kompetensi', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()

plot_filename_1 = 'gap_comparison_competency.png'
plt.savefig(plot_filename_1)
print(f"Visualisasi Kategori 1 (Kompetensi) disimpan sebagai {plot_filename_1}")

# Menampilkan pilihan terbaik:
print("\n[PILIHAN] Kompetensi Terbaik (Gap Terbesar):")
print(df_comp[['variable', 'difference']].head(1).to_markdown(index=False, numalign="left", stralign="left"))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Tentukan nama file yang akan diolah
file_name = "step2 pake benchmark.csv"
df_data = pd.read_csv(file_name)

# 1. Pastikan kolom skor adalah numerik
numeric_cols = ['final_match_rate', 'tgv_match_rate']
for col in numeric_cols:
    df_data[col] = pd.to_numeric(df_data[col], errors='coerce')

# 2. Cari Kandidat Terbaik (Berdasarkan Skor Final)
# Drop duplicates karena setiap karyawan muncul 3 kali (untuk 3 TGV)
df_ranked = df_data[['employee_id', 'final_match_rate']].drop_duplicates()
df_ranked = df_ranked.sort_values(by='final_match_rate', ascending=False)

top_candidate_id = df_ranked.iloc[0]['employee_id']
top_candidate_score = df_ranked.iloc[0]['final_match_rate']

# 3. Filter Data untuk Visualisasi
df_candidate = df_data[df_data['employee_id'] == top_candidate_id].copy()

# 4. Ambil Data Baseline Mentah dan Normalisasi untuk Visualisasi Benchmark
benchmark_scores_mentah = {
    'GTQ': df_data[df_data['tv_name'] == 'GTQ']['baseline_score'].iloc[0],
    'STO': df_data[df_data['tv_name'] == 'STO']['baseline_score'].iloc[0],
    'Papi_V': df_data[df_data['tv_name'] == 'Papi_V']['baseline_score'].iloc[0]
}

# Skala Max: GTQ (46), STO (5), Papi_V (9)
max_ranges = {'GTQ': 46, 'STO': 5, 'Papi_V': 9}

# Fungsi Normalisasi yang digunakan untuk Benchmark
def normalize_benchmark(tv_name, score):
    min_val = 1 if tv_name in ['STO', 'Papi_V'] else 9 # GTQ min 9
    max_val = max_ranges[tv_name]
    return (score - min_val) / (max_val - min_val) * 100

# Hitung Benchmark Normalized Score
bench_norm_scores = [
    normalize_benchmark('GTQ', benchmark_scores_mentah['GTQ']),
    normalize_benchmark('STO', benchmark_scores_mentah['STO']),
    normalize_benchmark('Papi_V', benchmark_scores_mentah['Papi_V'])
]

# --- Radar Chart Plotting ---
labels = df_candidate['tgv_name'].tolist()
cand_norm_scores = df_candidate['tgv_match_rate'].tolist()
final_labels = [f"{t} ({v})" for t, v in zip(labels, df_candidate['tv_name'].tolist())]

# Formatting the plot
num_vars = len(final_labels)
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]
cand_norm_scores += cand_norm_scores[:1]
bench_norm_scores += bench_norm_scores[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

ax.plot(angles, bench_norm_scores, color='blue', linewidth=2, linestyle='solid', label='Benchmark Profile (Rata-rata)')
ax.fill(angles, bench_norm_scores, color='blue', alpha=0.1)

ax.plot(angles, cand_norm_scores, color='red', linewidth=2, linestyle='solid', label=f'Top Candidate: {top_candidate_id}')
ax.fill(angles, cand_norm_scores, color='red', alpha=0.3)

# Customize the plot
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks(angles[:-1])
ax.set_xticklabels(final_labels)
ax.set_yticks(np.arange(0, 101, 25))
ax.set_yticklabels([f'{i}%' for i in np.arange(0, 101, 25)])
ax.set_ylim(0, 100)
ax.set_title(f'Kandidat Terbaik ({top_candidate_score:.2f}%) vs. Profil Ideal', size=14, y=1.1)
ax.legend(loc='lower right', bbox_to_anchor=(0.1, 0.1))

# Save the plot
plot_filename = 'radar_chart_step2_final.png'
plt.savefig(plot_filename)
print(f"Visualisasi Radar Chart telah disimpan sebagai {plot_filename}")

print(f"\nKandidat Terbaik (ID): {top_candidate_id}")
print(f"Final Match Score: {top_candidate_score:.2f}%")
```
